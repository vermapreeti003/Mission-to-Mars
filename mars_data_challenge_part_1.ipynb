```python
from splinter import Browser
from bs4 import BeautifulSoup as soup
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
```

```python
executable_path = { 'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
url_news='https://redplanetscience.com/'
browser.visit(url_news)
```

```python
html=browser.html
soup = soup(html,'html.parser')
```

```python
print(soup.prettify)
```

```python
#Scrapping the title text for each article
titles=soup.find_all('div',class_='content_title')

a=1
for title in titles:
    print(a,".",title.text)
    a=a+1
```

```python
#Scrapping the previews text for each article
previews = soup.find_all('div', class_='article_teaser_body')

b=1
for preview in previews:
    print(b,'.',preview.text)
    b=b+1
```

```python
news_list= []
news_dict = {}
item = soup.find_all('div',class_='list_text')
for t in item:
   
    title= t.find_all('div',class_='content_title')[0].text
    preview=t.find_all('div',class_='article_teaser_body')[0].text.strip()
    news_dict={'title': title,
               
              'preview': preview,
              }
    
    news_list.append(news_dict)
```

```python
print(news_list)
```

```python
# Export python list to JSON file
import json
json_mars = json.dumps(news_list)
json_mars
```

```python
# Save json as csv file
df=pd.read_json(json_mars)
df.to_csv('mars.csv',index=False)
```

```python
browser.quit()
```
